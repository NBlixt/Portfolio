# Brief Introduction

This notebook will investigate whether expression of specific pro-survival genes in the BCL-2 family correlates with cancer cell survival in response to treatment with inhibitors of said pro-survival factors.  Data from the [Genomics of Drug Sensitivity in Cancer (GDSC)](https://www.cancerrxgene.org/) and the [Cancer Dependency Map (DepMap)](https://depmap.org/portal/download/) will be used.  [Previous reports](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5990650/) have concluded that pro-survival BCL-2 family members are increased in expression in various cancer types.

## BCL-2 Family
The B-cell CLL/lymphoma 2 (BCL-2) family of proteins regulate a critical step in commitment to apoptosis by regulating permeabilization of the mitochondrial outer membrane (MOM).  Within this family are three classes:<br>
>1) Proapoptotic proteins that directly regulate permeabilization.<br>
>2) BCL-2 homology domain-3 (BH-3) proteins that directly or indirectly activate the pore-forming class members.<br>
>3) The pro-survival (anti-apoptotic) proteins that can inhibit the process at several steps.  This groups includes Bcl-2, Bcl-XL, and Mcl-1.<br><br>

## More About the Data
The GDSC data should include a list of cell lines and a list of drugs used in their experiemnts, as well as the response of each cell line to each drug.  Response will be measured as area under the curve (AUC), where a higher AUC means more survival/less effectiveness of the drug, and a lower AUC means less survival/more effectiveness of the drug.

To examine how expression of pro-survival factors is connected to drug response, I will be loading expression data from the DepMap dataset for cell lines shared between DepMap and GDSC.

### Gene Names Versus Protein Names
MCL-1 and BCL-2 have the same names for protein and gene, but the gene name doesn't have the hyphen.  However, BCL-XL is the protein name for the gene BCL2L1.  Throughout this notebook I will mostly likely use these names interchangeably.  

## Hypothesis
I predict that cell lines expressing high levels of one of the BCL-2 pro-survival factors will be more susceptible to drugs targeting that specific pro-survival factor and less susceptible to drugs targeting other pro-survival factors.

## Goals
My main goal is to construct plots that examine response to a specific drug relative to expression levels of the BCL-2 family members.

## Drugs Targeting BCL-2 Family Members
Find any drugs that may affect BCL-2, BCL-XL, or MCL-1, and display all information for those drugs.  Since there is sometimes inconsistencies in terms of hyphens, replace these elements before checking for the targets.  Also, the targets column appears to contain a list of items, so I will need to split the list.

```python
bcl2_drugs = []
for drug,targets in zip(drug_list.name, drug_list.targets):
    targets = targets.lower().replace("-", "").split(",")
    # Check if any of three genes are present in targets
    if any(gene in ["bcl2", "mcl1", "bclxl"] for gene in targets):
        bcl2_drugs.append(drug)
         
bcl2_drugs
```

```python
# Display all information for the three drugs
drug_list[drug_list.name.isin(bcl2_drugs)]
```

## Inhibitors of the BCL-2 Family
According the the information above, there are three drugs present in the drug list that will target BCL-2 family members.  Here is additional information on all three.

[Navitoclax](https://www.medchemexpress.com/navitoclax.html?src=google-product&gclid=EAIaIQobChMI2tGgtZvx4wIVGYzICh2QVwF-EAAYAiAAEgKlyPD_BwE) binds BCL-xL, and BCL-2 with similar $K_{i}$ values of 1 nM, but it does not appear to [inhibit MCL-1](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4669759/).  Navitoclax can induce thrombocytopenia as a result of inhibiting BCL-XL.  This prompted development of venetoclax, a BCL-2-selective inhibitor.

[Obatoclax Mesylate](https://www.tocris.com/products/obatoclax-mesylate_6539?gclid=EAIaIQobChMIr4X7_Y7v4wIViZ-fCh2J_gjzEAAYBCAAEgJThfD_BwE) is an antagonist of BCL-2 with a $K_{i}$ of 0.22 M.  Not reported to bind other family members (according to TW-37 link), but supposedly helps assist in overcoming MCL-1-mediated resistance to apoptosis.  The same website also states that Obatoclax binds to BCL-2, BCL-XL, and MCL-1.  According to [this paper](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5601122/), obatoclax did not affect protein levels of MCL-1, but it did disrupt MCL-1 and Bak interactions and promoted translocation of Bax into the mitochondria (both steps in Mitochondrial Outer Membrane Permeabilization (MOMP)), and [this paper](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2148320/) also supports a model in which obatoclax inhibits MCL-1.  Therefore, I will proceed with the conclusion that obatoclax will target BCL-2, BCL-XL, and MCL-1.


[TW-37](https://www.selleckchem.com/products/TW-37.html) is reported to inhibit BCL-2, BCL-XL, and MCL-1 with a [$K_{i}$ 0.29, 1.11, and 0.26 M](https://clincancerres.aacrjournals.org/content/13/7/2226), respectively.

Based on this, it looks like obatoclax mesylate and TW-37 will both affect all three BCL-2 family members, while navitoclax will target BCL-2 and BCL-XL but not MCL-1.

### Thoughts

The majority of myeloma cell lines treated with navitoclax appear to be relatively resistant (AUC between 0.9 and 1), but all cell lines were strongly inhibited (inhibition of an anti-apoptotic factor == more apoptosis and lower AUC) by obatoclax and less so by TW 37.  Why is there this disparate response to BCL-2 inhibitors?

#### Different Targets
Since navitoclax was the only drug of three that did not inhibit MCL-1, MCL-1 inhibition could be necessary for some/most myeloma cell lines to be efficiently induced to undergo apoptosis.  This could explain the different response to navitoclax, but not the two drugs that also target MCL-1.  In their case, perhaps the shift upwards in AUC for cell lines treated with TW-37 is merely due to different doses of each drug.  After all, the $K_{i}$ for both drugs is [similar](#BCL-2-Inhibitors); however, obatoclax treated doses were higher (0.06 - 16 M) than TW-37 (0.01 - 5 M).  Unfortunately, there isn't much I can glean from additional data if this truly is the main reason for the difference in response. 
  
#### Different Mutations or Expression
Related to the above explanations, the more resistant cell lines may have point mutations that alter their reponse to navitoclax only.  Or MCL-1 expression could be higher, due to various mutations or alterations in copy number.  So the higher the expression, the more sensitive the cell line would be to inhibition by obatoclax or TW 37.

#### Different Mechanisms
Lastly, aside from inhibition of MCL-1, there could be additional mechanisms behind how each drug affects apoptosis/survival of these cell lines. This/these additional mechanism(s) could be targeted more thoroughly by obatoclax, less so by TW-37, and even less by navitoclax.

## Clean and Combine Expression and Drug Response Data
I want to combine the expression and drug response data into a single data frame.  However, each drug of interest has a different number of cell lines that were used, and the expression data has different cell lines as well.  So, I will have to separate expression and response into three separate data frames, one for each drug.  This will make creating Bokeh plots later much easier.  In order to avoid repeating the cleaning and concatenation steps, I will make a function that will do this for me.

```python
def clean_concat_response_expression(expression_df, response_df, drug):
    """
    expression_df: The data frame containing gene expression information.
    
    response_df: The data frame containing drug response data.
    
    drug: the drug of interest that will be used to filter the response_df.
    
    For both data frames, the column with cell lines should not be the index,
    and it should be titled as 'cell_line.'
    
    This function will take a drug response dataset and a gene expression dataset,
    remove duplicates, filter so only shared cell lines betwen the two datasets
    are kept, and combine all the data into a single data frame (combined_df).
    
    return: combined_df
    """

    
    # Filter drug response by drug of interest
    drug_df = response_df.loc[response_df["drug_name"] == drug].copy()
    
    # Remove columns from drug resopnse except for cell_line, tissue_type, drug_name, and auc
    drug_df = drug_df.loc[:, ["cell_line", "tissue_type", "drug_name", "auc"]].copy()
    
    # Remove duplicates
    expression_df.drop_duplicates(subset="cell_line", inplace=True)
    drug_df.drop_duplicates(subset="cell_line", inplace=True)
    
    # Filter each df by the cell lines in the other
    drug_df = drug_df.loc[drug_df["cell_line"].isin(expression_df["cell_line"])].copy()
    expression_df = expression_df.loc[expression_df["cell_line"].isin(drug_df["cell_line"])].copy()
    
    # Concat the data frames
    combined_df = pd.concat([drug_df.set_index("cell_line"), 
           expression_df.set_index("cell_line")], axis=1, sort=True)
    
    # Set the index name to "cell_line"
    combined_df.index.rename("cell_line", inplace=True)
    
    return combined_df
```

# Plotting Response Versus Expression

There are two main goals for plotting.  First, I want to generate a scatter plot showing AUC relative to expression of a certain gene.  Second, I want to use linear regression to obtain a loose idea whether there is any relationship between expression of said gene and AUC.  

Plotting will be done with Bokeh again, and sci-kit learn (sklearn).  I'll design two separate functions for generating the graphs and generating the linear regression model.  Then I'll visualize a few plots and compare r-squared values.

## Examining All Tissue Types
To start, I will include all data from all tissue types to see if any overarching relationships are consistent.  Though, I expect this is not the case.

```python
def plot_linear_regression(x_values, y_values, bokeh_fig):
    """
    x_values: The plotted x_values in the bokeh_fig.
    
    y_values: The plotted y_values in the bokeh_fig.
    
    bokeh_fig: The Bokeh figure to which a linear regression line will be added.
    
    Takes a set of x-values, y-values, and the bokeh_fig with the plotted data.
    Calculates a linear regression line using the x-values and y-values, then
    adds the linear regression model to the plot as a red line.
    
    Calculates the r-squared value for the linear regression line and 
    adds this value as text to the bottom-right corner of the plot.
    
    return: The original Bokeh figure with an added linear regression line and text,
    and the r-squared value.
    """
    # Reshape the x_values for use in model
    x_values = np.array([x_values]).reshape(-1, 1)
    
    # Instantiate the model
    model = LinearRegression().fit(x_values, y_values)
    
    # Calculate the intercept and slope
    intercept = model.intercept_
    slope = model.coef_[0]
    
    # Create a regression line and add to the Bokeh figure
    regression_line = Slope(gradient=slope, y_intercept=intercept, line_color="red")
    bokeh_fig.add_layout(regression_line)
    
    # Determine and plot the R-squared value as text
    r2 = model.score(x_values, y_values)
    r2 = round(r2, 2)
    r2_text = Label(x=180, y=10, 
                x_units="screen", y_units="screen", 
                text="R^2 = " + str(r2), 
                render_mode="css",
                border_line_color="black",
                background_fill_color="white"
               )
    bokeh_fig.add_layout(r2_text)
    
    return bokeh_fig, r2
```

```python
def exp_response_bokeh(response_drug_df, gene):
    """
    response_drug_df: a data frame containing drug response and gene expression
    data made from clean_concat_response_expression().
    
    gene: the gene of interest for creating x-axis values.
    
    Creates a Bokeh scatter plot with gene expression on the x-axis and auc
    on the y-axis for the supplied data.
    
    return: the Bokeh plot
    
    """
    # Create the ColumnDataSource for supplying data to plots
    cds = ColumnDataSource(response_drug_df)
    
    # Interactive tools for displaying information when hovering over data
    hover_tools = [("Cell Line", "@cell_line"),
                   ("Tissue Type", "@tissue_type"),
                   ("AUC", "@auc"),
                   ("MCL-1", "@MCL1{0,000}"),
                   ("BCL-XL", "@BCL2L1{0,000}"),
                   ("BCL-2", "@BCL2{0,000}")
                  ]

    # Tools for maniuplating the plots
    select_tools = ["save", HoverTool(tooltips=hover_tools), "pan", "wheel_zoom", "reset"]

    # The actual figure area
    exp_response = figure(title="AUC Relative to " + gene,
                          x_axis_label=gene,
                          y_axis_label="AUC",
                          plot_width=350,
                          plot_height=350,
                          tools=select_tools)

    # The scatter plot data
    exp_response.scatter(source=cds, x=gene, y="auc", fill_color="DodgerBlue")
    
    # Create separate x_values and y_values for linear regression
    x_values = response_drug_df[gene]
    y_values = response_drug_df["auc"]
    
    # Add the linear regression model to the Bokeh plot and obtain r-squared
    exp_response, r2 = plot_linear_regression(x_values, y_values, exp_response)

    # Format the axes and labels
    for fig in [exp_response]:
        fig.xaxis.formatter = NumeralTickFormatter(format='0,000')
        label_size = "12pt"
        unit_size = "10pt"

        fig.title.align="center"
        fig.title.text_font_size = label_size
        fig.xaxis.major_label_text_font_size = unit_size
        fig.xaxis.axis_label_text_font_size = label_size
        fig.yaxis.major_label_text_font_size = unit_size
        fig.yaxis.axis_label_text_font_size = label_size
#         fig.xaxis.major_label_orientation = .7
        
    return exp_response, r2
```

```python
# Create dictionaries for storing the Bokeh plots
navitoclax_bokeh_dict = {}
obatoclax_bokeh_dict = {}
tw37_bokeh_dict = {}
for gene in ["BCL2", "BCL2L1", "MCL1"]:
    exp_response, r2 = exp_response_bokeh(nav_df, gene)
    navitoclax_bokeh_dict[gene] = exp_response
    
    exp_response, r2 = exp_response_bokeh(obato_df, gene)
    obatoclax_bokeh_dict[gene] = exp_response
    
    exp_response, r2 = exp_response_bokeh(tw37_df, gene)
    tw37_bokeh_dict[gene] = exp_response
```
